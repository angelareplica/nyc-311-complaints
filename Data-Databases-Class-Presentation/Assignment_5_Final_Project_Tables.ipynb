# Community Board 2
### Brooklyn Heights, DUMBO, Fulton Mall, Boerum Hill, Fort Greene, Brooklyn Navy Yard, Fulton Ferry, and Clinton Hill
Population: 143,328
https://censusreporter.org/profiles/79500US3604004-nyc-brooklyn-community-district-2-brooklyn-heights-fort-greene-puma-ny/

Number of HPD complaints filed since 2014: 18,985

```python
#Dataframes named based on Community Board
df2 = pd.read_csv("/Users/angelawang/Documents/nyc_housing/hpd_complaints_dumbo.csv")
#df2.head()
df2.shape
```

```python
# removing the .0 from the end of each zip
df2['zip'].fillna(0, inplace=True)
df2['zip'] = df2.zip.astype(np.int64)
#df2.head()
```

```python
df2['address'] = df2['housenumber'].astype('str') + " " + df2['streetname'] + " " + "Brooklyn, NY" + " " + df2['zip'].astype('str')
df2.head()
```

```python
df2.majorcategory.value_counts(normalize=True) * 100
```

```python
df2.majorcategory.value_counts()
```

```python
df2.code.value_counts()
```

```python
df2.minorcategory.value_counts()
```

```python
#df2.statusdescription.value_counts().head()
```

```python
#df2.bbl.isnull().value_counts()
```

```python
df2.majorcategory.isnull().sum() # 0 null values here
df2.code.isnull().sum() #628 nulls here
```

```python
dumbo_unsanitary_df = df2[df2.majorcategory == 'UNSANITARY CONDITION']
dumbo_unsanitary_df.head()
```

```python
dumbo_unsanitary_df[dumbo_unsanitary_df.code =='ACCUMULATION'].head()
```

```python
dumbo_unsanitary_df.code.isnull().sum() #628 nulls
dumbo_unsanitary_df.code.value_counts()
#"Accumulation" appears to mean garbage
```

```python
dumbo_unsanitary_df.code.value_counts(normalize=True)
```

```python
dumbo_unsanitary_df.minorcategory.value_counts()
# the 628 null values from "code" are mold
```

```python
dumbo_unsanitary_df.minorcategory.value_counts(normalize=True)
```

```python
dumbo_unsanitary_dummy_df = pd.get_dummies(dumbo_unsanitary_df, columns=['minorcategory', 'code'])
dumbo_unsanitary_dummy_df.dtypes
```

```python
#add a new column that organizes the code/minor categories
#dumbo_unsanitary_df.dtypes
dumbo_unsanitary_df['minorcategory_plus_code'] = dumbo_unsanitary_df['minorcategory'] + " - " + dumbo_unsanitary_df['code']
```

```python
dumbo_unsanitary_df['code'].fillna('None', inplace=True)
```

```python
# filter dataframe to look at the top unsanitary conditions
dumbo_unsanitary_dummy_df2 = dumbo_unsanitary_dummy_df[['complaintid', 'receiveddate', 'bbl', 'lat', 'lng', 'address', 'boroughid', 'majorcategory', 'minorcategory_plus_code', 'minorcategory_MOLD', 'minorcategory_GARBAGE/RECYCLING STORAGE', 'code_MICE', 'code_ROACHES', 'code_BED BUGS', 'minorcategory_SEWAGE']]

dumbo_unsanitary_dummy_df2
```

```python
dumbo_unsanitary_df.head()
dumbo_unsanitary_df[dumbo_unsanitary_df.minorcategory=="MOLD"]
```

```python
dumbo_unsanitary_dummy_df2.to_csv("dumbo_unsanitary_conditions_withdummies.csv", index=False)
```
